```python
#Import data tools
import pandas as pd
import numpy as np
import dash
from dash import dcc, html
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
#Extract info from excel
sheet_insumo = pd.read_excel("C:/Users/Cristian des/OneDrive/Documentos/Programación/Proyecto propios Data/Python/Dichter&neira/Prueba_Excel.xlsx",sheet_name ="Insumo")
sheet_info_tienda = pd.read_excel("C:/Users/Cristian des/OneDrive/Documentos/Programación/Proyecto propios Data/Python/Dichter&neira/Prueba_Excel.xlsx",sheet_name ="Info_Tiendas_mes")
sheet_info_productos = pd.read_excel("C:/Users/Cristian des/OneDrive/Documentos/Programación/Proyecto propios Data/Python/Dichter&neira/Prueba_Excel.xlsx",sheet_name ="Info_productos")
```

```python
#Exercise 1 "Praparación del insumo"
#Create new df from primary df
df = sheet_insumo
df1 = sheet_info_tienda
df2 = sheet_info_productos

#Append a new column to these news df's
df["Concat"] = df.apply(lambda row: str(row["Mes"])+row["Tienda"],axis = 1)
df1["Concat"] = df1.apply(lambda row: str(row["Mes"])+row["Tienda"],axis = 1)

#Merge df and df1 to "Concat" columns 
df_merge = pd.merge(df,df1, on="Concat",how="right")
df_merge1 = pd.merge(df_merge,df2, on="id_producto",how="right")
df_merge1["Concat"] = df_merge1.apply(lambda row: str(row["Mes_x"])+str(row["Región"])+str(row["Tienda_x"])+str(row["id_producto"]),axis = 1)
df_merge2 = df_merge1.drop(columns=["Mes_y","Tienda_y",])
df_merge2 = df_merge2.drop_duplicates(subset="Concat")
df_merge2 = df_merge2.dropna(subset=["Mes_x"])
```

```python
#Exercise 2 "Cálculo de variables de interés"
df_merge2["Sales Value Projected"] = df_merge2.apply(lambda row: row["cantidad"] * row["precio"] * row["wf"] * row["Fact_Colab"], axis=1)
df_merge2["Sales Projected Volume "] = df_merge2.apply(lambda row: row["cantidad"] * row["contenido_estandar"] * row["wf"] * row["Fact_Colab"], axis=1)
```

```python
#Exercise 3 "Tablas dinámicas"
#Sales Projected Value by zone and month
dinamictable_SaleValueProjectedByAreaByMonth = pd.DataFrame(df_merge2.groupby(["Mes_x","Región"])["Sales Value Projected"].sum().reset_index())
total_SalesValueByMonth = df_merge2.groupby(["Mes_x"])["Sales Value Projected"].sum().reset_index()

#Percentage sales Projected Value by zone and month 
merge_SalesValueByMonth = pd.merge(dinamictable_SaleValueProjectedByAreaByMonth,total_SalesValueByMonth,how= "right", on="Mes_x",suffixes=('','_total'))
merge_SalesValueByMonth["Percentage"] = pd.DataFrame(round((merge_SalesValueByMonth["Sales Value Projected"] / merge_SalesValueByMonth["Sales Value Projected_total"]) * 100,2))

#Sales Projected Volume by zone and month
dinamictable_SaleVolumeProjectedByAreaByMonth = pd.DataFrame(df_merge2.groupby(["Mes_x","Región"])["Sales Projected Volume "].sum().reset_index())
total_SalesVolumeByMonth = df_merge2.groupby(["Mes_x"])["Sales Projected Volume "].sum().reset_index()

#Percentage sales Projected Value by zone and month 
merge_SalesVolumeByMonth = pd.merge(dinamictable_SaleVolumeProjectedByAreaByMonth,total_SalesVolumeByMonth,how= "right", on="Mes_x",suffixes=('','_total'))
merge_SalesVolumeByMonth["Percentage"] = pd.DataFrame(round((merge_SalesVolumeByMonth["Sales Projected Volume "] / merge_SalesVolumeByMonth["Sales Projected Volume _total"]) * 100,2))

#print(dinamictable_SaleValueProjectedByAreaByMonth.head(4))
#print(merge_SalesValueByMonth[["Mes_x","Región","Percentage"]].head(4))
#print(dinamictable_SaleVolumeProjectedByAreaByMonth.head(4))
#print(merge_SalesVolumeByMonth[["Mes_x","Región","Percentage"]].head(4))
```

```python
#Exercise 4 "Important findings"
#Brand whith more sales projected value
BestSellingBrand = round(df_merge2.groupby(["marca"])["Sales Value Projected"].sum(),2)
BestSellingBrand = BestSellingBrand.sort_values(ascending= False)

#Find variety selling projected volume
TableBestSellingVariety = pd.merge(df_merge2,sheet_info_productos,how ="left", on= "id_producto", suffixes=("","info"))
BestSellingVariety = round(TableBestSellingVariety.groupby(["variedad"])["Sales Projected Volume "].sum(),2)
BestSellingVariety = pd.DataFrame(BestSellingVariety.sort_values(ascending= False))

#Primary targeting with almost percentaje 20%
TotalPrimaryTargeting = round(df_merge2["Sales Value Projected"].sum(),2)
DinamicPrimaryTargeting = round(df_merge2.groupby(["segmentacion_primaria"])["Sales Value Projected"].sum(),2)
Merge_DinamicPrimaryTargeting = pd.DataFrame(round((DinamicPrimaryTargeting / TotalPrimaryTargeting) * 100,2))

#Manufacture with greater decrement by month
ManufacturerMonth1filter = df_merge2[df_merge2['Mes_x'] == '2022-04-01']
ManufacturerWithGreaterDecrementMonth1 = round(ManufacturerMonth1filter.groupby(["fabricante"])["Sales Projected Volume "].sum())
ManufacturerWithGreaterDecrementMonth1 = ManufacturerWithGreaterDecrementMonth1.sort_values(ascending= False)
ManufacturerMonth2filter = df_merge2[df_merge2['Mes_x'] == '2022-07-01']
ManufacturerWithGreaterDecrementMonth2 = round(ManufacturerMonth2filter.groupby(["fabricante"])["Sales Projected Volume "].sum(),2)
ManufacturerWithGreaterDecrementMonth2 = round(ManufacturerWithGreaterDecrementMonth2.sort_values(ascending= False),2)
MergeManufactureMonths = pd.merge(ManufacturerWithGreaterDecrementMonth1,ManufacturerWithGreaterDecrementMonth2,how="inner",on="fabricante",suffixes=("Month1","Month2"))
MergeManufactureMonths["Profit/Lose"] = round(MergeManufactureMonths["Sales Projected Volume Month2"] - MergeManufactureMonths["Sales Projected Volume Month1"],)
DinamicManufacturerWithGreaterDecrementMonth = pd.DataFrame(MergeManufactureMonths["Profit/Lose"].sort_values())
DinamicManufacturerWithGreaterDecrementMonth = DinamicManufacturerWithGreaterDecrementMonth.head(5)
DinamicManufacturerWithGreaterDecrementMonth.reset_index(inplace=True)

#Sub brand with greater decrement by month
SubBrandGreaterDecrementMonth1 = df_merge2[(df_merge2['Mes_x'] == '2022-04-01')&(df_merge2['fabricante'] == 'Barraza')]
SubBrandWithGreaterDecrementMonth1 = round(SubBrandGreaterDecrementMonth1.groupby(["id_producto"])["Sales Projected Volume "].sum(),2)
SubBrandWithGreaterDecrementMonth1 = round(SubBrandWithGreaterDecrementMonth1.sort_values(ascending= False),2)
SubBrandGreaterDecrementMonth2 = df_merge2[(df_merge2['Mes_x'] == '2022-07-01')&(df_merge2['fabricante'] == 'Barraza')]
SubBrandWithGreaterDecrementMonth2 = round(SubBrandGreaterDecrementMonth2.groupby(["id_producto"])["Sales Projected Volume "].sum(),2)
SubBrandWithGreaterDecrementMonth2 = round(SubBrandWithGreaterDecrementMonth2.sort_values(ascending= False),2)
MergeSubBrandMonths = pd.merge(SubBrandWithGreaterDecrementMonth1,SubBrandWithGreaterDecrementMonth2,how="inner",on="id_producto",suffixes=("Month1","Month2"))
MergeSubBrandMonths["Profit/Lose"] = round(MergeSubBrandMonths["Sales Projected Volume Month2"] - MergeSubBrandMonths["Sales Projected Volume Month1"],)
SheetInfoFilter = df[["id_producto","descripcion"]]
DinamicSubBrandMonths = pd.DataFrame(round(MergeSubBrandMonths["Profit/Lose"].sort_values(),2))
MergeDinamicWithInfoProducts = pd.merge(DinamicSubBrandMonths,SheetInfoFilter,how="inner",on="id_producto")
MergeDinamicWithInfoProducts = MergeDinamicWithInfoProducts.drop_duplicates(subset="descripcion")
MergeDinamicWithInfoProducts.reset_index(inplace=True)

#Size market with the greatest project value and volume
SizeMarketGreat = df_merge2[df_merge2['Tamaño_tienda'] == 'Grd']
SizeMarketGreatProjectedValue = round(SizeMarketGreat.groupby(["Mes_x"])["Sales Value Projected"].sum(),2)
SizeMarketMedium = df_merge2[df_merge2['Tamaño_tienda'] == 'Med']
SizeMarketMediumProjectedValue = round(SizeMarketMedium.groupby(["Mes_x"])["Sales Value Projected"].sum(),2)
MergeCompareSizeMarket = pd.merge(SizeMarketGreatProjectedValue,SizeMarketMediumProjectedValue,how="inner",on="Mes_x",suffixes=(" Great Market"," Medium Market"))
MergeCompareSizeMarket["Compare"] = round(MergeCompareSizeMarket["Sales Value Projected Great Market"] - MergeCompareSizeMarket["Sales Value Projected Medium Market"],)
MergeCompareSizeMarket["Yes/No"] = MergeCompareSizeMarket['Compare'].apply(lambda x: 'Is bigger than Medium Market' if x > 0 else 'Isnt bigger than Med Market')
DinamicCompareSizeMarket = pd.DataFrame(MergeCompareSizeMarket[["Compare","Yes/No"]])

#What is the region whit greates projected value and volume
DinamicSalesValueAndVolumeProjectedPerZone = pd.DataFrame(df_merge2.groupby(["Mes_x","Región"])[["Sales Value Projected","Sales Projected Volume "]].sum())

#What is the brands with greates projected value
DinamicSalesValueProjectedPerBrand = pd.DataFrame(df_merge2.groupby(["marca"])["Sales Value Projected"].sum())
DinamicSalesValueProjectedPerBrand = pd.DataFrame(DinamicSalesValueProjectedPerBrand.sort_values(by="Sales Value Projected",ascending=False))
DinamicSalesValueProjectedPerBrand = DinamicSalesValueProjectedPerBrand.head()
DinamicSalesValueProjectedPerBrand.reset_index(inplace=True)

#What is the brands with greates projected volume
DinamicSalesVolumeProjectedPerBrand = pd.DataFrame(df_merge2.groupby(["marca"])["Sales Projected Volume "].sum())
DinamicSalesVolumeProjectedPerBrand = DinamicSalesVolumeProjectedPerBrand.sort_values(by="Sales Projected Volume ",ascending=False)
DinamicSalesVolumeProjectedPerBrand = DinamicSalesVolumeProjectedPerBrand.head()
DinamicSalesVolumeProjectedPerBrand.reset_index(inplace=True)

#Projected value sales by company size
ProjectedValueByMonth = pd.DataFrame(df_merge2.groupby(["Mes_x","Tamaño_tienda"])["Sales Value Projected"].sum())
ProjectedValueByMonth.reset_index(inplace=True)

#Projected sales value and volume by month and zone
ProjectedSalesValueVolumeByMonthZone = pd.DataFrame(df_merge2.groupby(["Mes_x","Región"])[["Sales Value Projected","Sales Projected Volume "]].sum())
ProjectedSalesValueVolumeByMonthZone.reset_index(inplace=True)

#10 febrero 1997

#print(ProjectedSalesValueVolumeByMonthZone)
#print(ProjectedValueByMonth)
#print(DinamicSalesVolumeProjectedPerBrand)
#print(DinamicSalesValueProjectedPerBrand)
#print(DinamicSalesValueAndVolumeProjectedPerZone)
#print(DinamicCompareSizeMarket)
#print(MergeDinamicWithInfoProducts)
#print(DinamicManufacturerWithGreaterDecrementMonth.head(5))
#print(Merge_DinamicPrimaryTargeting)
#print(BestSellingVariety)
```
